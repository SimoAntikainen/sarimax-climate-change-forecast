```python
from sklearn.metrics import mean_squared_error

#https://stackoverflow.com/questions/6682784/reducing-number-of-plot-ticks

X = maunaloaFilled.CO2filled.dropna()#hist_splice.data_mean_global
size = int(len(X) * 0.95)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot

#pyplot.plot(test)
#pyplot.plot(predictions, color='red')
#pyplot.locator_params(nbins=6)
#pyplot.figure(figsize=(20,10))
#pyplot.show()
```

## Seasonal Decomposition

```python
res = sm.tsa.seasonal_decompose(anomalyrange.anomaly,freq=12)
fig = res.plot()
fig.set_figheight(8)
fig.set_figwidth(15)
plt.show()
```

```python
res = sm.tsa.seasonal_decompose(co2range.CO2filled,freq=12)
fig = res.plot()
fig.set_figheight(8)
fig.set_figwidth(15)
plt.show()
```

## Checking Stationarity (Trend)

```python
res = sm.tsa.adfuller(anomalyrange.anomaly,regression='ct')
print('p-value:{}'.format(res[1]))
#has a unit root and is non-stationary
```

```python
res = sm.tsa.adfuller(co2range.CO2filled,regression='ct')
print('p-value:{}'.format(res[1]))
# does not have a unit root and is stationary
```
