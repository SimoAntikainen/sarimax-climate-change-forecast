```python
#co2historical['03-Jan-1880 00:00:00':'15-Dec-2014 00:00:00']
hist_splice = co2historical['15-Jan-1880 00:00:00':'15-Dec-2014 00:00:00']
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,7))
data_mean_global = hist_splice.data_mean_global.plot(ax=ax)
plt.xlabel('Date')
plt.ylabel('C02 ppm')
plt.title('Historical CO2 levels')
plt.show()
```

```python
model = ARIMA(hist_splice.data_mean_global, order=(5,1,0))
model_fit = model.fit(disp=0)
```

```python
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
```

```python
from sklearn.metrics import mean_squared_error

#https://stackoverflow.com/questions/6682784/reducing-number-of-plot-ticks

X = hist_splice.data_mean_global
size = int(len(X) * 0.95)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.locator_params(nbins=6)
pyplot.figure(figsize=(20,10))
pyplot.show()
```
